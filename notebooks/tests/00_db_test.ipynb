Hier wird getestet, ob eine Verbindung zur Datenbank aufgebaut werden kann.
Wenn die Ausgabe 1 ist, konnte die Verbindung hergestellt werden.

```python
from mph.db import get_engine
from sqlalchemy import text

engine = get_engine()
with engine.connect() as con:
    print(con.execute(text("select 1")).scalar())
```
